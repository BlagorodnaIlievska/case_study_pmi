# Checking assumtions about the features

1. MARKET is unique per REGION

```python
#checking if MARKET is unique per REGION

#create a dataframe of region1 only
region1=df_train[df_train['REGION']=='REGION1']
#take only the numeric part of the category
region1_market=region1['MARKET'].str.slice(start=6, stop=len(region1['MARKET']), step=1)
#create an array of all unique markets per region 1
region1_market=region1_market.unique()

#create a dataframe of region2 only
region2=df_train[df_train['REGION']=='REGION2']
#take only the numeric part of the category
region2_market=region2['MARKET'].str.slice(start=6, stop=len(region2['MARKET']), step=1)
#create an array of all unique markets per region 2
region2_market=region2_market.unique()

#create a dataframe of region3 only
region3=df_train[df_train['REGION']=='REGION3']
#take only the numeric part of the category
region3_market=region3['MARKET'].str.slice(start=6, stop=len(region3['MARKET']), step=1)
#create an array of all unique markets per region 2
region3_market=region3_market.unique()

#create a dataframe of region4 only
region4=df_train[df_train['REGION']=='REGION4']
#take only the numeric part of the category
region4_market=region4['MARKET'].str.slice(start=6, stop=len(region4['MARKET']), step=1)
#create an array of all unique markets per region 4
region4_market=region4_market.unique()

if np.intersect1d(region1_market, region2_market).size == 0:
    print('Markets in Region 1 do not appear in Markets of Region 2')
else:
    print('Markets in Region 1 appear in Markets of Region 2')
if np.intersect1d(region1_market, region3_market).size == 0:
    print('Markets in Region 1 do not appear in Markets of Region 3')
else: 
    print('Markets in Region 1 appear in Markets of Region 3')
if np.intersect1d(region1_market, region4_market).size == 0:
    print('Markets in Region 1 do not appear in Markets of Region 4')
else:
    print('Markets in Region 1 appear in Markets of Region 4')
if np.intersect1d(region2_market, region3_market).size == 0:
    print('Markets in Region 2 do not appear in Markets of Region 3')
else:
     print('Markets in Region 2  appear in Markets of Region 3')
if np.intersect1d(region2_market, region4_market).size == 0:
     print('Markets in Region 2 do not appear in Markets of Region 4')
else:
    print('Markets in Region 2 appear in Markets of Region 4')
if np.intersect1d(region3_market, region4_market).size == 0:
    print('Markets in Region 3 do not appear in Markets of Region 4')
else:
    print('Markets in Region 3 appear in Markets of Region 4')
```

The first assumption is confirmed.

2. BRM, BRANDONMARKET, MARKETEDBRAND are the same columns

```python
#check if these columns are the same

#take only the numeric part of BRM
brm=df_train['BRM'].str.slice(start=3, stop=len(df_train['BRM']), step=1)
#take only the numeric part of BRANDONMARKET
brand_on_market=df_train['BRANDONMARKET'].str.slice(start=13, stop=len(df_train['BRANDONMARKET']), step=1)
#take only the numeric part of MARKETEDBRAND
marketed_brand=df_train['MARKETEDBRAND'].str.slice(start=13, stop=len(df_train['MARKETEDBRAND']), step=1)

if brm.equals(brand_on_market):
    print('BRM is equal to BRANDONMARKET')
else:
    print('BRM is not equal to BRANDONMARKET')
    print((brm==brand_on_market).value_counts())
if brm.equals(marketed_brand):
    print('BRM is equal to MARKETEDBRAND')
else:
    print ('BRM is not equal to MARKETEDBRAND')
    print((brm==marketed_brand).value_counts())
if brand_on_market.equals(marketed_brand):
    print('BRANDONMARKET is equal to MARKETEDBRAND')
else:
    print('BRANDONMARKET is equal to MARKETEDBRAND')
    print((brand_on_market==marketed_brand).value_counts())

#they are not completely the same - but very similar
```

# Explore some of the categorical features

```python
def column_frequency(df, column):
    """This is a function that counts the number of category repretitions in a given feature.
        It returns a dataframe.
    """
    
    df = df.astype(str).groupby(column, as_index=False).size()
    df = df.reset_index()
    df = df.rename(columns={0: 'Counts'})
    df = df.sort_values('Counts', ascending=False)
    return df

def plot_category(df, column):
    df_merged=column_frequency(df,[column,'Successful'])
    df_merged = df_merged.rename(columns={'Counts':'Number of category repetitions'})
    plt.figure(1, figsize=(32, 10))
    plt.ylabel('',fontsize=20)
    plt.xlabel('',fontsize=20)
    plt.xticks(rotation=90, fontsize=14)
    plt.title(column+' '+'category frequencies', fontsize=20)
    sns.barplot(x=df_merged[column],y=df_merged['Number of category repetitions'],hue=df_merged['Successful'])
```

```python
df_train['SPECIALFLAVOR']=df_train['SPECIALFLAVOR'].fillna('NOFLAVOR')
plot_category(df_train, 'SPECIALFLAVOR')
```

```python
plot_category(df_train, 'REGION')
```

Check the date column

```python
months=df_train.groupby('NPLLAUNCHDATE', as_index=False).agg({'Market_Share':'mean'})
months=months.rename(columns={'Market_Share':'Mean_Market_Share'})

f, ax = plt.subplots(figsize=(22, 10))
g = sns.pointplot(x=months['NPLLAUNCHDATE'], y=months['Mean_Market_Share'], data=months)
plt.xticks(rotation=90);
plt.ylabel('',fontsize=20)
plt.xlabel('',fontsize=20)
plt.xticks(rotation=90, fontsize=10)
plt.title('Mean Market Share over the launch dates', fontsize=20)
```

```python
#create separate columns for the month and the year
df_train['LAUNCHYEAR']=df_train['NPLLAUNCHDATE'].astype(str).str[:4]
df_train['LAUNCHMONTH']=df_train['NPLLAUNCHDATE'].astype(str).str[-2:]

months=df_train.groupby('LAUNCHMONTH', as_index=False).agg({'Market_Share':'mean'})
months=months.rename(columns={'Market_Share':'Mean_Market_Share'})

f, ax = plt.subplots(figsize=(15, 8))
g = sns.pointplot(x=months['LAUNCHMONTH'], y=months['Mean_Market_Share'], data=months)
plt.xticks(rotation=90, fontsize=10);
plt.title('Mean Market Share for each of the months', fontsize=20)
```
