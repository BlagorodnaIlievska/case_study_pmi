```python
# Data Science
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression

pd.set_option('display.max_columns', 100)
```

```python
df_train = pd.read_excel('D:/krk-datascientist-master/case2/use_case_data.xlsx')
df_test = pd.read_excel('D:/krk-datascientist-master/case2/score_data.xlsx')
test_offset = df_test.shape[0]
print(test_offset)
```

```python
df = pd.concat([df_train, df_test], axis=0, sort=True)
df.tail(40)
```

```python
#Merge REGION and MARKET into one feature
df['REGION_MARKET']=df['REGION'].astype(str)+'_'+df['MARKET'].astype(str)

#Encode Date into Months
df['LAUNCHMONTH']=df['NPLLAUNCHDATE'].astype(str).str[-2:]

#Add Target variable
df['Successful']=np.where(df['Market_Share']<0.007, 0, 1)

#SPECIALFLAVOR - we fill the missing values with a 'no flavor' category under the assumption that the product did not have a flavor 
df['SPECIALFLAVOR']=df['SPECIALFLAVOR'].fillna('NOFLAVOR')

#TIPCOLOR - we fill the missing values with a 'no tipcolor' category under the assumption that the product did not have a color
df['TIPCOLOR']=df['TIPCOLOR'].fillna('NOCOLOR')

#TIPCOLOR - we fill the missing values with a 'no tipcolor' category under the assumption that the product did not have a color
df['RTYPE']=df['RTYPE'].fillna('NORTYPE')

#TIPCOLOR - we fill the missing values with a 'no tipcolor' category under the assumption that the product did not have a color
df['ITEMSHAPE']=df['ITEMSHAPE'].fillna('NOITEMSHAPE')
```

```python
df=df.drop(['BRANDSUBFAMILYGROUPING','BRMID','ISREPLACEMENT', 'LATESTPERIODINDEX', 
            'REGION', 'MARKET','NPLLAUNCHDATE','NPLLAUNCHYEAR','Market_Share'],axis=1)
```

```python
# Transform ITEMSCODE to categorical
df['ITEMSCODE']=df['ITEMSCODE'].astype('category')

# Transform the Dataframe to a Dataframe with dummy encoding
df=pd.get_dummies(df)

df.tail(40)
```

```python
df_train = df[:-test_offset]
df_train.tail(10)
```

```python
df_test = df[-test_offset:]
df_test.head(test_offset)
```

```python
y_train = df_train['Successful'].values
X_train = df_train.drop(['Successful'], axis=1)
```

```python
X_test = df_test.drop(['Successful'], axis=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# For example

```python
#the parameters are defined 
y_pred = LogisticRegression(C=1.200, penalty='l1',class_weight='balanced', random_state=1234, solver= 'liblinear').fit(X_train, y_train).predict(X_test)
```

```python
df_test['Successful'] = y_pred
```

```python
writer = pd.ExcelWriter('D:/krk-datascientist-master/case2/new_score_data_labels.xlsx')
df_test.to_excel(writer,'Sheet1')
writer.save()
```
